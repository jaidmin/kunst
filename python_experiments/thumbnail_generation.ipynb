```python
import PIL.Image
import numpy as np
```

```python
cat1 = PIL.Image.open("normal.jpg")
cat2 = PIL.Image.open("styled.jpg")
```

```python
cat1.size
```

```python
mask =np.triu(np.ones((1000,1000))*255).astype('float32')
```

```python
mask = np.rot90(mask)
```

```python
maskimage = PIL.Image.fromarray(mask.astype('uint8'),mode='L')
```

```python
maskimage = maskimage.resize((500,332))
```

```python
average = np.ones((50,50))/2500
from scipy.ndimage import convolve
mask = np.array(maskimage)
mask = convolve(mask,average,)#output=np.array((10,10),dtype='float32')
```

```python
mask = mask.astype('uint8')
maskimage = PIL.Image.fromarray(mask,mode='L')
```

```python
cat1 = PIL.Image.open("normal.jpg")
cat2 = PIL.Image.open("styled.jpg")
```

```python
cat1.paste(cat2,mask=maskimage)
```
